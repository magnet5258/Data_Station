- 추출 :

    - 열단위 : df1[] / df1[[]]

```python
# 데이터 정렬 (오름차순)
df1.sort_values(by = '정지시간2')
```

```python
# 데이터 추출 (행단위, iloc : index location)
df1.iloc[0:100] # 0번째부터 100번째 전(99번째)까지의 100개 데이터를 추출
```

```python
# 데이터 추출 (열단위)
df1[['메세지', '정지시간2', '작업구분', '작업자']]
# Series : 순서와 값으로 구성 * index - value / df1[]
# DataFrame 형태로 데이터 추출 * DataFrame : index - colum - value / df1[[]]
```

```python
# 데이터 필터 (특정 조건을 만족하는 데이터를 추출, loc : location)
cond1 = (df1['정지시간2'] >= 600)
df1.loc[cond1]
```

```python
# 두 조건을 모두 만족하는 데이터만 추출
cond1 = (df1['정지시간2'] >= 600)
cond2 = (df1['작업구분'] == '야간')
df1.loc[cond1 & cond2] # 두 조건을 모두 만족하는 데이터를 추출 (& : and)
```

```python
# 두 조건 중 하나라도 만족하는 데이터를 추출
cond1 = (df1['정지시간2'] >= 600)
cond2 = (df1['메세지'] == '초기화 Error')
df1.loc[cond1 | cond2] # 두 조건 중 하나라도 만족하는 데이터를 추출 (| : or)
```

```python
# 데이터 요약 (pivot_table)
df1.pivot_table(index = '메세지', values = '정지시간2', aggfunc = 'mean')
```

# 기계학습

**PDA (Predictive Data Analysis) : 예측적 데이터 분석**

```python
df1 = pd.read_csv('preprocessing_data.csv')
df1.head()
```

```python
df1.info()
```

```python
# 데이터 핸들링
df1.isnull().sum() # 결측값 확인
```

```python
# Wafer 폐기여부를 판별하는 모델을 생성
df1['target_binom'].value_counts()
```

```python
df1.columns
```

```python
Y = df1['target_binom']
# drop() : 데이터 프레임에서 불필요한 항목을 제거
X = df1.drop(columns = ['Target', 'Error_message', 'target_binom', 'Chamber_Route',
                        'Unnamed: 0', 'Ox_Chamber', 'process', 'type', 'No_Die',
                        'Unnamed: 0_x', 'photo_soft_Chamber', 'process 2',
                        'Unnamed: 0_y', 'Etching_Chamber', 'Process 3', 'Chamber_Num',
                        'process4'])
```

```python
# 숫자데이터만 선택
numeric_list = X.describe().columns.tolist()
X1 = X[numeric_list]
```

```python
# 학습데이터와 검증데이터 분할
from sklearn.model_selection import train_test_split
```

```python
# 학습데이터와 검증데이터를 8:2 비율로 분할
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, random_state = 1234)
# random_state : 무작위 추출을 원하는 index로 맞춰 추출
print(X_train.shape) # shape : 데이터의 행과 열 개수를 추출
print(X_test.shape)
```

```python
# 학습 수행
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train) # fit : 학습을 수행
```

```python
# 평가
from sklearn.metrics import accuracy_score
```

```python
# 정확도 : 전체 데이터 중 정확하게 분류한 데이터의 수 (0 ~ 1)
Y_train_pred = model.predict(X_train)
accuracy_score(Y_train, Y_train_pred)
```

```python
Y_test_pred = model.predict(X_test)
accuracy_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision_score(Y_test, Y_test_pred)
```

```python
recall_score(Y_test, Y_test_pred)
```

```python
f1_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_test_pred))
```
